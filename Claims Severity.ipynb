# Explatory data analysis

```python
for column in df.columns:
    # Show only categorical variables unique values
    if str(df[column].dtype.name) == 'object':
        print("col:",column,"unique values:",df[column].unique())
```

```python
# Exploring for null values
Flag = False
for c in df.columns:
    # Show only categorical variables unique values
    count = df[column].isnull().sum()
    if count > 0:
        print("col:",column, "has",count,"null values")
        Flag = True
if not Flag:
    print("No null values have been found")
```

## Only numerical values could be digested by the machine learning algorithms so we have to encode the characters to numbers assigning them from 0 to their count.
### A = 0, B = 1, ... Z = 25

```python
# iterate through each column
for column in df.columns:
    # check if column contains numerical data
    if df[column].dtype == 'float64' or df[column].dtype == 'int64':
        continue
    # create a dictionary to map unique values to their count
    value_counts = df[column].value_counts().to_dict()
    # create a new column with encoded values
    encoded_column = df[column].apply(lambda x: list(value_counts.keys()).index(x))
    # replace the original column with the encoded column
    df[column] = encoded_column
```

```python
df
```

```python
X = df.drop("loss",axis=1)
y = df['loss']
```

```python
X
```

```python
y
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error

from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import HistGradientBoostingRegressor
```

```python
def predict_mae(c,X,y,p=False):
    print(c)
    y_preds = c.predict(X)
    if p:
        print(mean_absolute_error(y,y_preds))
    return mean_absolute_error(y,y_preds)
```

```python
len(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# Training algorithms individually and finding out their Mean Absolute Error (the lower the better)

```python
MLP_R = MLPRegressor()
MLP_R.fit(X_train,y_train)
predict_mae(MLP_R,X_test,y_test)
```

```python
KNN_R = KNeighborsRegressor()
KNN_R.fit(X_train,y_train)
predict_mae(KNN_R,X_test,y_test)
```

```python
Bagging_R = BaggingRegressor()
Bagging_R.fit(X_train,y_train)
predict_mae(Bagging_R,X_test,y_test)
```

```python
HGB_R = HistGradientBoostingRegressor()
HGB_R.fit(X_train,y_train)
predict_mae(HGB_R,X_test,y_test)
```

```python
import mustafapy
```

```python
from mustafapy import mustafa
m = mustafa(atype='reg',precision=2,hyperstep=5,maxattrs=2,selection=['KNeighborsRegressor','BaggingRegressor', 'HistGradientBoostingRegressor'])
hypers_dictionary = m.hypers()
m.hypers()
```

```python
regressors = [KNeighborsRegressor(),BaggingRegressor(),HistGradientBoostingRegressor()]
for r in regressors:
    print(hypers_dictionary[r.__class__])
```

```python
for reg in regressors:
    parameters = hypers_dictionary[reg.__class__]

    reg_GS = RandomizedSearchCV(reg,
                          parameters,
                          cv=None,
                          scoring='neg_mean_absolute_error',
                           n_jobs=-1,
                            n_iter=10,
                            verbose=3
                          )

    reg_GS.fit(X, y)
    predict_mae(reg_GS.best_estimator_,X_test,y_test,p=True)
```
